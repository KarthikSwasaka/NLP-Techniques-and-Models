```python
from scipy import spatial
cosine_similarity = lambda vec1, vec2: - spatial.distance.cosine(vec1, vec2)
```

```python
nlp.vocab['king']
```

```python
king = nlp.vocab['king'].vector
man = nlp.vocab['man'].vector
woman = nlp.vocab['woman'].vector
```

```python
new_vector = king - man + woman
```

```python
new_vector
```

```python
compueted_similarities = []
for word in nlp.vocab:
    if word.has_vector:
        if word.is_lower:
            if word.is_alpha:
                similarity = cosine_similarity(new_vector, word.vector)
                compueted_similarities.append((word, similarity))
```

```python
compueted_similarities = sorted(compueted_similarities, key = lambda item:-item[1])
```

```python
print([t[0].text for t in compueted_similarities[:10]])
```
